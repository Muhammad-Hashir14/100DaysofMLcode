```python
import pandas as pd
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/fifa_data.csv')
```

```python
gas = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv')
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
```

```python
fig, ax = plt.subplots(figsize = (8, 5), dpi = 100)
ax.plot(x,y,'b^--',label = '2x')

x2 = np.arange(0, 4.5, 0.5)
ax.plot(x2[:6], x2[:6] **2 , 'r', label = 'x^2')
ax.plot(x2[5:], x2[5:] ** 2, 'r--')

plt.title('My first Graph',fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X VALUES')
plt.ylabel('Y VALUES')
plt.legend()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

fig, ax = plt.subplots(figsize = (8,5))
bars = plt.bar(labels, values)

patterns = ['/','o','*']
for i in bars:
    i.set_hatch(patterns.pop(0))
plt.savefig('patterns.png', dpi = 300)
plt.show()
```

## Real World Examples

```python
fig, ax = plt.subplots(figsize = (8,5))
plt.plot(gas.Year, gas['USA'],'y.-', label = 'USA')
plt.plot(gas.Year,gas.Canada,'g.-',label = 'Canada')
plt.plot(gas.Year, gas['South Korea'],'b.-',label = 'South Korea')
plt.plot(gas.Year, gas.Australia, 'r.-',label = 'Australia')
plt.title('Gas Price Over Time')
plt.xlabel('Time')
plt.ylabel('price')
plt.legend()
plt.savefig('gas prices',dpi = 300)
```

```python
right = fifa[fifa['Preferred Foot'] == 'Right'].count()[1]
left = fifa[fifa['Preferred Foot'] == 'Left'].count()[1]
labels = ['Right','Left']
colors = ['#abcdef', '#aabbcc']
plt.pie([right,left], labels = labels,colors=colors,autopct='.%2f')
plt.title('FIFA players foot preference')
plt.savefig('pie.png',dpi=300)
```

```python
fig, ax = plt.subplots(figsize = (8,5),dpi = 100)
plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]
weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)
plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels = label, explode = explode,pctdistance=0.8 ,autopct='.%2f')
```

```python
plt.figure(figsize=(5,8), dpi=100)
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    box.set(color='#4286f4', linewidth=2)
    box.set(facecolor = '#e0e0e0' )
```
