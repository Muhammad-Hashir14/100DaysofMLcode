```python
def get_var(df, var_name):
    globals()[var_name] = df
    return df

def tweak_autos(vehicles):
    return (vehicles
     [cols]
     #create var
     .pipe(get_var, 'df3')
     .assign(cylinders = vehicles['cylinders'].fillna(0).astype('int8'),
            disp = vehicles.displ.fillna(0).astype('float16'),
            drive = vehicles.drive.fillna('Other').astype('category'),
            automatic = vehicles.trany.str.contains('Auto'),
            speed = vehicles.trany.str.extract(r'(\d)+').fillna(20).astype('int8'))
     #debug pipe
     .pipe(lambda df: display(df) or df)       
     .astype({'highway08': 'int8', 'city08': 'int16', 'comb08': 'int16', 'fuelCost08': 'int16',
             'range': 'int16', 'year': 'int16', 'make': 'category'})
     .drop(columns = ['trany'])
)

tweak_autos(vehicles)
```
